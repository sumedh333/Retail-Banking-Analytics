# Cleaning Tables

```python
def default_cleaning(df):
    df = df.copy()

    for col in df.columns:
        if df[col].dtype == "object":
            df[col] = (
                df[col]
                .astype(str)
                .str.strip()
                .str.lower()
                .replace(["", "nan", "none"], np.nan)
            )
    return df
customer = default_cleaning(customer)
account = default_cleaning(account)
transaction = default_cleaning(transaction)
loan = default_cleaning(loan)
card = default_cleaning(card)
feedback = default_cleaning(feedback)

customer = default_cleaning(customer)
account = default_cleaning(account)
transaction = default_cleaning(transaction)
loan = default_cleaning(loan)
card = default_cleaning(card)
feedback = default_cleaning(feedback)
```

# Exploratory Data Analysis

## 1. Which age groups maintain the highest average account balances across account types?

```python
# join customer + account
df = customer.merge(account, on="customer_id", how="inner")

# create age groups
df["age_group"] = pd.cut(
    df["age"],
    bins=[18,25,35,50,65,120],
    labels=["18-25","26-35","36-50","51-65","65+"]
)

# compute avg balance per segment
seg = df.groupby(["age_group","account_type"])["account_balance"].mean().reset_index()

plt.figure(figsize=(9,6))
sns.barplot(data=seg, x="age_group", y="account_balance", hue="account_type")

plt.title("Average Account Balance by Age Group and Account Type")
plt.ylabel("Avg Account Balance")
plt.xlabel("Age Group")

# move legend to right
plt.legend(title="Account Type", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.tight_layout()
plt.show()
```

### Insights
- Customers aged 36–50 maintain the highest average account balances across both current and savings accounts, indicating stronger earning and saving capacity in mid-career groups.
- Savings accounts consistently show higher balances than current accounts in every age segment, suggesting customers prefer savings products for holding funds.
- The 65+ group shows stable balances but at a slightly lower level, indicating reduced inflow after retirement.


## 2. How does transaction frequency vary between savings and current account holders?

```python
# join account + transaction
df = account.merge(transaction, on="customer_id")

# count transactions
count_tx = df.groupby("account_type")["transaction_id"].count().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=count_tx, x="account_type", y="transaction_id")

#  Y-axis label
plt.ylabel("Number of Transactions")

#  X-axis label
plt.xlabel("Account Type")

# title
plt.title("Transaction Frequency by Account Type")
plt.tight_layout()
plt.show()
```

### Insights
- Savings account holders make more transactions than current account holders.  
- This indicates that savings accounts are used more actively for routine banking activity.


## 3. How does total transaction value vary across days of the week?

```python
# extract day of week
transaction["day_of_week"] = transaction["transaction_date"].dt.day_name()

# convert transaction amount to lakhs
transaction["transaction_amount_lakhs"] = transaction["transaction_amount"] / 100000

# total value by day of week (in lakhs)
dow = (
    transaction.groupby("day_of_week")["transaction_amount_lakhs"]
    .sum()
    .reset_index()
)

# reorder days
order = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
dow = dow.set_index("day_of_week").loc[order].reset_index()

# plot
plt.figure(figsize=(12,6))
sns.barplot(data=dow, x="day_of_week", y="transaction_amount_lakhs", ci=None)

plt.title("Transaction Value by Day of the Week (in Lakhs)")
plt.xlabel("Day of Week")
plt.ylabel("Total Transaction Amount (Lakhs)")
plt.xticks(rotation=45)
plt.show()
```

### Insights
- Transaction values remain consistently high throughout the week, with only minor variation between weekdays and weekends.  
- Sunday records the highest total transaction value, indicating increased customer activity at the end of the week.  
- Wednesday shows the lowest transaction value, but the difference is small, suggesting evenly distributed activity across all days.


## 4. Do high-value customers (top 10% by balance) transact differently than others?

```python
# join account + transaction
df = account.merge(transaction, on="customer_id", how="inner")

# find 90th percentile balance (top customers)
threshold = df["account_balance"].quantile(0.90)

# flag high-value customers
df["customer_segment"] = np.where(df["account_balance"] >= threshold, "High-Value", "Regular")

# compare transaction behaviour
seg = df.groupby("customer_segment")["transaction_amount"].agg(["count","mean","sum"]).reset_index()

# convert sum to lakhs
seg["sum_lakhs"] = seg["sum"] / 100000

plt.figure(figsize=(10,6))
sns.barplot(data=seg, x="customer_segment", y="sum_lakhs")
plt.title("High-Value vs Regular Customers – Total Transaction Value (Lakhs)")
plt.ylabel("Total Transaction Value (Lakhs)")
plt.show()

```

### Insights
- Regular customers contribute a significantly higher total transaction value than high-value customers, indicating that high balances do not necessarily translate into higher spending.
- High-value customers appear less transaction-active, suggesting they may use accounts primarily for holding funds rather than frequent usage.
- This gap highlights an opportunity to design targeted engagement programs to increase transaction activity among high-value customers.


## 5. How does average transaction amount differ across account types?

```python
# join account + transaction
df = account.merge(transaction, on="customer_id", how="inner")

# compute average transaction amount by account type
grp = df.groupby("account_type")["transaction_amount"].mean().reset_index()

# plot
sns.barplot(data=grp, x="account_type", y="transaction_amount")
plt.title("Average Transaction Amount by Account Type")
plt.show()
```

### Insights
- The average transaction amount is nearly identical for current and savings account holders.  
- This indicates that account type has little influence on the size of individual transactions.


## 6. Which cities show the highest average transaction value per customer?

```python
# join customer + transaction
df = customer.merge(transaction, on="customer_id", how="inner")

# compute total transaction amount per customer
cust_tx = df.groupby(["customer_id", "city"])["transaction_amount"].sum().reset_index()

# now compute average per customer for each city
city_avg = cust_tx.groupby("city")["transaction_amount"].mean().reset_index()

city_avg = city_avg.sort_values("transaction_amount", ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(data=city_avg.head(10), y="city", x="transaction_amount")
plt.title("Top 10 Cities by Average Transaction Value per Customer")
plt.xlabel("Average Transaction Value")
plt.ylabel("City")
plt.show()

```

### Insights
- All top cities show relatively high average transaction values, but Kansas City, Mesa, and Detroit stand slightly above others, indicating stronger per-customer spending behaviour.
- The differences between cities are small, suggesting that high-value transactions are not concentrated in just one or two regions but are distributed across several mid-sized cities.
- These cities may represent attractive targets for premium banking products since customers already exhibit higher-than-average transaction activity.


## 7. Which cities show higher borrowing-to-spending behavior (Loan-to-Transaction Ratio)?

```python
# loan amount by city
loan_city = loan.merge(customer[["customer_id","city"]], on="customer_id")
loan_amt = loan_city.groupby("city")["loan_amount"].sum().reset_index()
loan_amt.rename(columns={"loan_amount":"total_loan"}, inplace=True)

# transaction amount by city
tx_city = transaction.merge(customer[["customer_id","city"]], on="customer_id")
tx_amt = tx_city.groupby("city")["transaction_amount"].sum().reset_index()
tx_amt.rename(columns={"transaction_amount":"total_spend"}, inplace=True)

# merge
ratio = loan_amt.merge(tx_amt, on="city", how="inner")

# compute ratio
ratio["loan_to_spend_ratio"] = ratio["total_loan"] / ratio["total_spend"]

# sort & take top 10
top10_ratio = ratio.sort_values("loan_to_spend_ratio", ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(data=top10_ratio, y="city", x="loan_to_spend_ratio")
plt.title("Top Cities by Loan-to-Transaction Ratio")
plt.xlabel("Loan to Spending Ratio")
plt.ylabel("City")
plt.show()

```

### Insights
- Charlotte shows the highest loan-to-transaction ratio, meaning customers there take significantly larger loans relative to their transaction activity. This indicates strong demand for credit products but comparatively lower spending behaviour.
- Colorado Springs, Atlanta, and Boston also display elevated ratios, pointing to regions where customers rely more on borrowing than on daily banking activity.
- All top cities maintain consistently high ratios (around 10–12), suggesting a cluster of markets where credit appetite is strong. These cities offer strategic potential for loan-focused campaigns, risk monitoring, and targeted cross-selling.


## 8. Which feedback types (complaint, suggestion, praise) are most common among loan customers?

```python
# loan customer flag
loan_flag = loan[["customer_id"]].drop_duplicates().assign(has_loan=True)

# join loan flag into feedback table
df = feedback.merge(loan_flag, on="customer_id", how="left")
df["has_loan"] = df["has_loan"].fillna(False)

# filter only loan customers
loan_feedback = df[df["has_loan"] == True]

# count feedback types
ftype = loan_feedback["feedback_type"].value_counts().reset_index()
ftype.columns = ["feedback_type", "count"]

# plot
plt.figure(figsize=(8,5))
sns.barplot(data=ftype, x="feedback_type", y="count", ci=None)
plt.title("Most Common Feedback Types Among Loan Customers")
plt.xlabel("Feedback Type")
plt.ylabel("Count")
plt.show()
```

### Insights
- All three feedback types appear at similar volumes, indicating a balanced distribution of customer responses among loan holders.  
- Suggestions and complaints occur slightly more frequently than praise, implying customers tend to report issues or recommend improvements more often than express positive feedback.


## 9. What is the yearly trend in loan approvals and rejections?

```python
# extract year from approval/rejection date
loan["year"] = loan["approval_rejection_date"].dt.year

# group by year and loan status
loan_trend = loan.groupby(["year", "loan_status"])["loan_id"].count().reset_index()

# plot
plt.figure(figsize=(10,6))
sns.barplot(data=loan_trend, x="year", y="loan_id", hue="loan_status")

plt.title("Yearly Trend in Loan Approvals and Rejections")
plt.ylabel("Number of Loans")
plt.xlabel("Year")
plt.xticks(rotation=45)

plt.legend(title="Loan Status")
plt.tight_layout()
plt.show()
```

### Insights
- Loan approvals increased steadily from 2020 to 2022, indicating rising demand or more favorable lending conditions.  
- Rejections remained relatively stable across the three years, showing no major change in risk profile.  
- Closed loans also saw a slight upward trend, suggesting consistent repayment or account closure activity.

## 15. Does credit card utilization vary across different loan term lengths?

### Insights
- Utilization ratios are fairly similar across all loan term groups, showing no strong relationship between loan duration and credit card usage.  
- Customers with loan terms of 37–48 months show slightly higher utilization, indicating marginally greater credit usage in this segment.  
- The 49–60 month group has the lowest utilization, suggesting more conservative credit behavior among customers with longer loan commitments.
